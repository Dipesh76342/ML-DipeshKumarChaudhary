```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
def encode_categoricals(df):
    le = LabelEncoder()
    for col in df.select_dtypes(include=['object', 'category']).columns:
        if col != 'Diagnosis':
            df[col] = le.fit_transform(df[col].astype(str))
    return df

df = encode_categoricals(df)
```

```python
df['Diagnosis'] = LabelEncoder().fit_transform(df['Diagnosis'])
```

```python
X = df.drop('Diagnosis', axis=1)
y = df['Diagnosis']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print('Train shape:', X_train.shape)
print('Test shape:', X_test.shape)
```

```python
# Before Tuning
rf = RandomForestClassifier(n_estimators=150, max_depth=4, random_state=42)
rf.fit(X_train, y_train)

svc = SVC(kernel='rbf', C=0.68, gamma=0.05, probability=True, random_state=42)
svc.fit(X_train, y_train)

xgb = XGBClassifier(n_estimators=10, max_depth=3, learning_rate=0.1, use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

def evaluate_model(model, X_tr, y_tr, X_te, y_te, name):
    y_tr_pred = model.predict(X_tr)
    y_te_pred = model.predict(X_te)
    y_te_proba = model.predict_proba(X_te)[:,1] if hasattr(model, 'predict_proba') else None
    print(f'--- {name} ---')
    print('Train Accuracy:', accuracy_score(y_tr, y_tr_pred))
    print('Test Accuracy:', accuracy_score(y_te, y_te_pred))
    print('Test Precision:', precision_score(y_te, y_te_pred))
    print('Test Recall:', recall_score(y_te, y_te_pred))
    print('Test F1:', f1_score(y_te, y_te_pred))
    if y_te_proba is not None:
        print('Test ROC AUC:', roc_auc_score(y_te, y_te_proba))
    print(classification_report(y_te, y_te_pred))
    print('\n')
    return y_te_pred, y_te_proba

# to avoid multiple executions
results = []
for model, name in zip([svc, rf, xgb], ['SVM', 'Random Forest', 'XGBoost']):
    y_pred, y_proba = evaluate_model(model, X_train, y_train, X_test, y_test, name)
    results.append((y_pred, y_proba))

y_pred_svc, y_proba_svc = results[0]
y_pred_rf, y_proba_rf = results[1]
y_pred_xgb, y_proba_xgb = results[2]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


scoring = ['accuracy', 'f1', 'roc_auc']
models = {'SVM': svc, 'Random Forest': rf, 'XGBoost': xgb}

print("\n=== Cross-Validation Results ===")
for name, model in models.items():
    print(f"\n--- {name} CV ---")
    scores = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    for metric in scoring:
        mean_score = scores[f'test_{metric}'].mean()
        std_score = scores[f'test_{metric}'].std()
        print(f"{metric.upper()} - Mean: {mean_score:.4f}, Std: {std_score:.4f}")
```

```python

# After tuning models
# Random Forest
rf_params = {'n_estimators': [27,60], 'max_depth': [2, 3], 'min_samples_leaf': [3, 5, 10], 'class_weight': ['balanced']}
gs_rf = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=5, scoring='f1', n_jobs=-1)
gs_rf.fit(X_train, y_train)
print('Random Forest Best Params:', gs_rf.best_params_)

# XGBoost 
xgb_params = {'n_estimators': [20, 30], 'max_depth': [2, 3], 'learning_rate': [0.01, 0.05]}
gs_xgb = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), xgb_params, cv=5, scoring='recall', n_jobs=-1)
gs_xgb.fit(X_train, y_train)
print('XGBoost Best Params:', gs_xgb.best_params_)


# SVM 
svm_params = {'C': [0.1, 0.4], 'kernel': ['rbf', 'linear'], 'gamma': ['scale', 0.01, 0.05, 0.1]}
gs_svm = GridSearchCV(SVC(probability=True, class_weight='balanced', random_state=42), svm_params, cv=5, scoring='accuracy', n_jobs=-1)
gs_svm.fit(X_train, y_train)
print('SVM Best Params:', gs_svm.best_params_)

best_rf = gs_rf.best_estimator_
best_xgb = gs_xgb.best_estimator_
best_svm = gs_svm.best_estimator_


```

```python
tuned_rf = best_rf.fit(X_train, y_train)
tuned_xgb = best_xgb.fit(X_train, y_train)
tuned_svm = best_svm.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, confusion_matrix, classification_report, roc_curve

def evaluate_model(model, X_tr, y_tr, X_te, y_te, name):
    y_tr_pred = model.predict(X_tr)
    y_te_pred = model.predict(X_te)
    y_te_proba = model.predict_proba(X_te)[:,1] if hasattr(model, 'predict_proba') else None
    print(f'--- {name} ---')
    print('Train Accuracy:', accuracy_score(y_tr, y_tr_pred))
    print('Test Accuracy:', accuracy_score(y_te, y_te_pred))
    print('Test Precision:', precision_score(y_te, y_te_pred))
    print('Test Recall:', recall_score(y_te, y_te_pred))
    print('Test F1:', f1_score(y_te, y_te_pred))
    if y_te_proba is not None:
        print('Test ROC AUC:', roc_auc_score(y_te, y_te_proba))
    print(classification_report(y_te, y_te_pred))
    print('\n')
    return y_te_pred, y_te_proba

results = []
for model, name in zip([tuned_svm, tuned_rf, tuned_xgb], ['SVM', 'Random Forest', 'XGBoost']):
    y_pred, y_proba = evaluate_model(model, X_train, y_train, X_test, y_test, name)
    results.append((y_pred, y_proba))

y_pred_svm, y_proba_svm = results[0]
y_pred_rf, y_proba_rf = results[1]
y_pred_xgb, y_proba_xgb = results[2]


fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, y_pred, name in zip(axes, [y_pred_svm, y_pred_rf, y_pred_xgb], ['SVM', 'Random Forest', 'XGBoost']):
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', ax=ax)
    ax.set_title(f'Confusion Matrix: {name}')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
plt.tight_layout()
plt.show()


plt.figure(figsize=(8,6))
for y_proba, name in zip([y_proba_svm, y_proba_rf, y_proba_xgb], ['SVM', 'Random Forest', 'XGBoost']):
    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        plt.plot(fpr, tpr, label=name)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend()
plt.show()

```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


scoring = ['accuracy', 'f1', 'roc_auc']
models = {'SVM': svc, 'Random Forest': rf, 'XGBoost': xgb}

print("\n=== Cross-Validation Results ===")
for name, model in models.items():
    print(f"\n--- {name} CV ---")
    scores = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    for metric in scoring:
        mean_score = scores[f'test_{metric}'].mean()
        std_score = scores[f'test_{metric}'].std()
        print(f"{metric.upper()} - Mean: {mean_score:.4f}, Std: {std_score:.4f}")
```
